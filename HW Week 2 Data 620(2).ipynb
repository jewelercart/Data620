### Import the required modules or python library

```python
import networkx as network
import matplotlib.pyplot as plt
```

### Let's create an empty graph

```python
# Let's create an empty graph 
G = network.Graph()
```

### Now let's defined nodes and edged of the data. 

```python
nodes = ["Andre", "Beverly", "Carol", "Diane", "Ed", "Fernando", "Garth", "Heather", "Ike", "Jane"]

edges = [("Andre", "Beverly"), ("Andre", "Carol"), ("Andre", "Diane"), ("Beverly", "Diane"), ("Beverly", "Ed"),
         ("Beverly", "Garth"), ("Beverly", "Fernando"), ("Carol", "Diane"), ("Carol", "Fernando"), ("Diane", "Fernando"),
         ("Diane", "Garth"), ("Diane", "Ed"), ("Ed", "Garth"), ("Ed", "Fernando"), ("Fernando", "Garth"),
         ("Garth", "Heather"), ("Heather", "Ike"), ("Ike", "Jane")]
```

### Let's add nodes and edges to the empty graph created above 

```python
G.add_nodes_from(nodes)
G.add_edges_from(edges)
```

### Let's plot the graph 

```python
network.draw(G, with_labels=True, node_size=1000, node_color="lightgreen", font_size=12)
plt.title("Network Graph: Krackhardt kite")
plt.show()
```
